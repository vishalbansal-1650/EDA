```python
#Create new variable as “No_of_30_plus_DPD = No_of_30_59_DPD+No_of_90_DPD+ No_of_60_89_DPD”, “No_of_60_plus_DPD = No_of_90_DPD+ No_of_60_89_DPD” in cust_data
```

```python
Cust_data['No_of_30_plus_DPD'] = Cust_data['No_of_30_59_DPD']+Cust_data['No_of_90_DPD']+Cust_data['No_of_60_89_DPD']
```

```python
#Perform below joins between cust_data & cust_demo.
#a. Create data set “cust_leftjoin” using Left Join
#b. Create data set “cust_rightjoin” using right Join
#c. Create data set “cust_innterjoin” using inner Join
#d. Create data set “cust_fulljoin” using full Join
```

```python
#a
cust_leftjoin=pd.merge(Cust_data,Cust_Demo,how='left',on='ID')
```

```python
#b
cust_rightjoin=pd.merge(Cust_data,Cust_Demo, how='right',on='ID')
```

```python
#c
cust_innterjoin=pd.merge(Cust_data,Cust_Demo,how='inner',on='ID')
```

```python
cust_fulljoin=pd.merge(Cust_data,Cust_Demo,how='outer',on='ID')
```

```python
#Create summary report with below column by combination of Gender, Serious_delinquency.
#a. Number of customers
#b. Percentage of customers
#c. Average Revolving utilization
#d. Standard deviation with in the utilization
#e. Average monthly income
#f. Maximum Monthly income
#g. Standard deviation with in the income
#h. Average Age
#i. Average Dependents
```

```python
cust_leftjoin.head()
```

```python
grouped = cust_leftjoin.groupby(['Gender','SeriousDlqin2yrs'])
```

```python
customer_count = grouped['ID'].count()
```

```python
customer_count.sum()
```

```python
customer_percent=grouped['ID'].count()/customer_count.sum() * 100
```

```python
Rev_util_mean = grouped['RevolvingUtilization'].mean()
```

```python
Rev_util_std = grouped['RevolvingUtilization'].std()
```

```python
MonthlyIncome_mean = grouped['MonthlyIncome'].mean()
```

```python
MonthlyIncome_max = grouped['MonthlyIncome'].max()
```

```python
MonthlyIncome_std = grouped['MonthlyIncome'].std()
```

```python
age_mean = grouped['age'].mean()
```

```python
No_of_dependents_mean = grouped['No_of_dependents'].mean()
```

```python
pd.concat([customer_count,customer_percent,Rev_util_mean,Rev_util_std,MonthlyIncome_mean,MonthlyIncome_max,MonthlyIncome_std,age_mean],
          keys=['cust_count','cust_%','Avg_RevUtil','StdRevUtil','Avg_MonthlyIncome','Max_MonthlyIncome','Std_MonthlyIncome','Avg_age'],
         axis=1)
```
